```python
import re

def calc_acc():
    labels1 = []
    labels2 = []
    with open('data/log_query.txt') as f:
        for s_line in f:
            label = []
            l = s_line.split(',')
            for l_ in l[1:-1]:
                label.append(int(l_))
            labels1.append(label)
    with open('data/log_pred.txt') as f:
        for s_line in f:
            l = s_line.split(',')
            label = []
            for l_ in l[1:-1]:
                label.append(int(l_))
            labels2.append(label)

    cnt1 = 0
    cnt2 = 0
    for label1, label2 in zip(labels1, labels2):
        for i,j in zip(label1, label2):
            if i%2 == j:
                cnt1 += 1
            cnt2+= 1

    print('accuracy: ', cnt1 /cnt2)
calc_acc()
```
