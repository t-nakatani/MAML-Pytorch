```python
import pandas as pd
spt = pd.read_csv('../support_set.csv')
qry = pd.read_csv('../query_set.csv')
```

```python
pres = list(spt['pre'].unique())

import numpy as np
pres = np.array(pres)
np.random.seed(0)
np.random.shuffle(pres)
```

```python
dic = {}
for i, pre in  enumerate(pres):
    if i < 40:
        dic[pre] = 'test'
    else:
        dic[pre] = str((i-40) // 40)
```

```python
pres_spt = list(spt['pre'] )
pres_qry = list(qry['pre'] )
```

```python
spt['cv'] = [dic[pre] for pre in pres_spt]
```

```python
spt.groupby('cv').count()
```

```python
qry['cv'] = [dic[pre] for pre in pres_qry]
qry.groupby('cv').count()
```

```python
spt.to_csv('suppot_cv2.csv', index = False)
qry.to_csv('query_cv2.csv', index = False)
```
